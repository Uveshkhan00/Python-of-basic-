2. What is an interpreter in Python ?

   - An interpreter in Python is a software program that reads and executes Python code directly, line by line. Unlike a compiler, which translates the entire program into machine code before execution, the Python interpreter translates and runs the code on the fly. This process is what makes Python an interpreted language.

3. What are pre-defined keywords in Python ?
  - Pre-defined keywords in Python are reserved words that have a specific meaning and purpose in the Python language. These keywords are part of Python's syntax, and they cannot be used as identifiers (e.g., variable, function, or class names) because they are reserved for programming constructs.



4.Can keywords be used as variable names ?
  - No, keywords cannot be used as variable names in Python. They are reserved words in the Python programming language and are intended for specific syntactical or structural purposes. Using a keyword as a variable name will result in a SyntaxError.

5. What is mutability in Python ?
  - Mutability in Python refers to whether an object’s state (its data or content) can be changed after it is created. Python objects are categorized as either mutable or immutable based on this characteristic.



6. Why are lists mutable, but tuples are immutable ?
  - The difference in mutability between lists and tuples in Python is a deliberate design choice based on their intended use cases. Here's an explanation:

Lists: Mutable
Reason: Lists are designed to be dynamic and flexible, allowing modification of their content after creation. This includes operations like adding, removing, or updating elements.
Use Case: Lists are used when you need a data structure to manage collections that can change over time.
Implementation: Internally, lists are implemented with a resizable array. Modifications to the list do not require creating a new object; the existing list is updated in place.

9 What is type casting in Python ?
  - Type casting in Python refers to the process of converting one data type to another. It allows programmers to explicitly change the type of a variable, which can be useful when you need to perform operations that require specific types.

Python provides two types of type casting:

Implicit Type Casting (automatic by Python)
Explicit Type Casting (manual by the programmer)


10. What is the difference between implicit and explicit type casting?
   - Type casting in Python involves converting one data type to another. It can be categorized into two types: Implicit and Explicit. Here's a detailed comparison:

Feature	Implicit Type Casting	Explicit Type Casting
Definition	Automatically performed by Python when it is safe.	Performed manually by the programmer using built-in functions.
Control	Python controls the type conversion.	The programmer has full control over the conversion.
Risk of Data Loss	No risk; Python ensures safe conversions.	Risk of data loss if conversion is incompatible or truncates data.
Use Case	Applied when types are compatible (e.g., int to float).	Used to force conversion between different types (e.g., str to int).
Error Handling	Python handles errors and avoids unsafe conversions.	Programmer must ensure the conversion is valid to avoid runtime errors.
Examples	Integer to Float, Float to Complex, etc.	String to Integer, Float to Integer, etc.

12. How does the elif statement work?

13. What is the difference between for and while loops?
  - Both for and while loops in Python are used for iteration, allowing you to repeatedly execute a block of code. However, they differ in how they work and when they are used.

Feature	for Loop	while Loop
Control Mechanism	Iterates over a sequence (e.g., list, range, string).	Repeats as long as a condition evaluates to True.
Use Case	When the number of iterations is known or the iteration is over a fixed sequence.	When the number of iterations is not predetermined, and you rely on a condition to terminate the loop.
Syntax	for variable in sequence:	while condition:
Termination	Stops when the sequence is fully traversed.	Stops when the condition becomes False.
Suitability	Best for iterating over items in a collection or range.	Best for indefinite loops or loops controlled by user input or calculations.
Common Use Cases	Iterating over lists, strings, dictionaries, or ranges.	Repeatedly prompting user input, waiting for specific conditions.
Risk of Infinite Loop	Low, as it depends on a sequence.	Higher, if the condition is never met or improperly updated.


14. Describe a scenario where a while loop is more suitable than a for loop.
   - Scenario Where a while Loop Is More Suitable Than a for Loop
A while loop is more suitable than a for loop when the number of iterations is not predetermined and depends on a dynamic condition.

Example Scenario: User Login System
Imagine a login system where a user is prompted to enter their password. The program should keep asking for the password until the user enters the correct one. Since the number of attempts is unknown, a while loop is ideal for this situation.

4.Write a program that checks if a given word is a Python keyword?

```python
# Import the keyword library
import keyword

# Get user input
word = input("Enter a word: ")

# Check if the word is a Python keyword
if keyword.iskeyword(word):
    print(f"'{word}' is a Python keyword.")
else:
    print(f"'{word}' is not a Python keyword.")
```
